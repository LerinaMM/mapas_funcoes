# Estilos de mapas

```python
# Mapa em estilo Preto e Branco
folium.Map(
    location = [-23.5489, -46.6388],
    tiles = 'Stamen Toner'
)
```

```python
# Mapa em estilo Satélite
folium.Map(
    location = [-23.5489, -46.6388],
    tiles = 'Stamen Terrain'
)
```

```python
# Mapa em estilo Satélite
folium.Map(
    location = [-23.5489, -46.6388],
    tiles = 'Stamen Terrain',
    zoom_start = 3 # escolhe o zoom que você quer que apareça no mapa, quanto menor, mais longe
)
```

```python
# Quando quero marcar um ponto dentro do mapa
folium.Map(
    location = [-23.5489, -46.6388],
    tiles = 'Stamen Terrain',
)

folium.Marker(
    [-23.5503099,-46.6342009],
    popup = 'Praça da Sé',
    tooltip = 'Click Aqui'
).add_to(mapa)
```

```python
mapa
```

```python
# Quando quero marcar um ponto de outra cor dentro do mapa
folium.Map(
    location = [-23.5489, -46.6388],
    tiles = 'Stamen Terrain',
)

folium.Marker(
    [-23.5503099,-46.6342009],
    popup = 'Praça da Sé',
    tooltip = 'Click Aqui',
    icon = folium.Icon(color = 'green')
).add_to(mapa)
```

```python
mapa
```

```python
# Quando quero colocar um circulo de raio dentro do mapa
folium.Map(
    location = [-23.5489, -46.6388],
    tiles = 'Stamen Terrain',
)

folium.CircleMarker(
    [-23.5503099,-46.6342009],
    radius = 50,
    color = '#3186cc',
    fill = True,
    fill_collor = '#3186cc'
).add_to(mapa)
```

```python
mapa
```

```python
#Achar a latitude e longitude em qualquer ponto do mapa, voce clica no ponto que voce quer e fornece o ponto de latitude e longitude no mapa

mapa.add_child(folium.LatLngPopup())
```

```python
#Importar outras bibliotecas

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings('ignore')
```

Inserindo o dataframe de pontos turisticos de São Paulo

```python
Dados = pd.read_excel('pontos_turisticos.xlsx')
```

```python
Dados.head()
```

```python
#Aqui vamos marcar todos os pontos turisticos do dataset que carregamos - pontos turisticos.xlsx

mapa = folium.Map(location = [-23.5489,-46.6388])
for local, lat, lon in zip(Dados.LOCAL, Dados.LATITUDE.values, Dados.LONGITUDE.values):
    folium.Marker([lat,lon], tooltip=f"{local}").add_to(mapa)
```
